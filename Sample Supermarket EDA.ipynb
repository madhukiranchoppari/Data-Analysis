## Handling Outliers in Data Frame

```python
#cheking outlier in dataframe using function for quantitive columns 
df1.describe()
```

```python
#checking the product which as unitprice as 6783
df1[df1['Unit_Price']>=6783]
```

from above function and plot we can understand that sales,disount,profit,unit_price and shipping_cost as outliers

### 1.Handling Outlier in sales column

```python
#outlier handling using iqr(inter quartile range) method for sales column
sq1= np.percentile(df1['Sales'],25)
sq3=np.percentile(df1['Sales'],75)
siqr=sq3-sq1
print(siqr)
#setting lower limit for sales column
slow=sq1-1.5*siqr
print(slow)
#setting higher limit for sales columns
shigh=sq3+1.5*siqr
print(shigh)
```

```python
#capping the outlier in sales colum with low and high values
df1['Sales']=np.where(df1['Sales']<slow,slow,np.where(df1['Sales']>shigh,shigh,df1['Sales']))
```

```python
#checking for outliers in sales after capping
plt.figure(figsize=(5,3))
sns.boxplot(df1['Sales'])
plt.show()
```

from above plot we can see there are zero outliers in sales column in dataframe

### 2.Handling outlier in Discount column

```python
#outlier handling using iqr(inter quartile range) method for discount column
dq1= np.percentile(df1['Discount'],25)
dq3=np.percentile(df1['Discount'],75)
diqr=dq3-dq1
print(diqr)
#setting lower limit for sales column
dlow=dq1-1.5*diqr
print(dlow)
#setting higher limit for sales columns
dhigh=dq3+1.5*diqr
print(dhigh)
```
